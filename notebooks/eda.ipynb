```python
# Caminho para o dataset no Google Drive
dataset_path = '/content/drive/MyDrive/Projetos/projeto-credito/data/application_train.csv'

# Ler o dataset
import pandas as pd
df = pd.read_csv(dataset_path)

# Verificar os dados
df.head()
```

```python
# Exibir informações básicas do DataFrame
df.info()

# Estatísticas descritivas
df.describe()

# Verificar valores ausentes
missing_values = df.isnull().sum()

# Contagem de valores em variáveis categóricas
categorical_cols = df.select_dtypes(include='object').columns
for col in categorical_cols:
    print(f'{col}:')
    print(df[col].value_counts())
    print()

# Verificar correlação entre as variáveis numéricas
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
# Exibir informações básicas do DataFrame
df.info()

# Estatísticas descritivas
df.describe()

# Verificar valores ausentes
missing_values = df.isnull().sum()

# Contagem de valores em variáveis categóricas
categorical_cols = df.select_dtypes(include='object').columns
for col in categorical_cols:
    print(f'{col}:')
    print(df[col].value_counts())
    print()

# Verificar correlação entre as variáveis numéricas
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numerical features for correlation analysis
numerical_df = df.select_dtypes(include=['number'])  # Select numerical columns only

plt.figure(figsize=(12, 8))
sns.heatmap(numerical_df.corr(), annot=False, cmap='coolwarm')  # Calculate correlation on numerical data
plt.show()
```
